```python
# Import Numpy for calculations and matplotlib for charting
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a list from start 0 and end 5 with step 0.1
x_axis = np.arange(0, 5, 0.1)
x_axis
```

```python
# Creates an exponential series of values which we can chart
e_x = [np.exp(x) for x in x_axis]
e_x
```

```python
# Plot function
plt.plot(x_axis, e_x)

# Show function (shows the graph in this case)
plt.show()
```

```python
# Graphing a sine and cosine together

x_axis2 = np.arange(0, 6, 0.1)

sin = np.sin(x_axis2)
cos = np.cos(x_axis2)

plt.plot(x_axis2, sin)
plt.plot(x_axis2, cos)

plt.show()
```

```python
# Activity 1: Plotting New Jersey Temperatures

# Fahrenheit degrees list
f = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]

# List comprehension to change f into c
c = [((i-32)*(5/9)) for i in f]

# Days arange (13 days as temperature values)
days = np.arange(1, 13, 1)

# Plotting days vs C and days vs F
plt.plot(days, c)
plt.plot(days, f)

# Show the graphs
plt.show()
```

```python
# Set x axis and variables
x_axis3 = np.arange(0, 10, 0.1)
sin = np.sin(x_axis3)
cos = np.cos(x_axis3)

# Plot sine and cosine
plt.plot(x_axis3, sin)
plt.plot(x_axis3, cos)

# Draw a horizontal line with 25% transparency
plt.hlines(0, 0, 10, alpha = 0.25)
```

```python
# Plot with modifications for a better, sleeker image
sine_handle = plt.plot(x_axis3, sin, marker = "o", color = "blue", label = "Sine")
cosine_handle = plt.plot(x_axis3, cos, marker = "^", color = "red", label = "Cosine")

# Locks the legend into a chosen place
plt.legend(loc = "lower right")

#plt.savefig("../Images/lineConfig.png")
plt.show()
```

```python
%matplotlib notebook
```

```python
# Activity 2: Modifying the data for better looks
f_handle = plt.plot(days, f, marker = "+", color = "b", label = "Fahrenheit")
c_handle = plt.plot(days, c, marker = "s", color = "r", label = "Celsius")

# Locks the legend into a chosen place
# plt.legend(handler = [f, c], loc = "best")
plt.legend(loc = "upper right")

# Adds labels and header
plt.xlabel("Months")
plt.ylabel("Degrees")
plt.title("Temperature in NJ")

# Saves the graph as .png
plt.savefig("lineConfig.png")
plt.show()
```

```python
# Adding aesthetics in Python
# Activity 3: Roller Coaster

# Roller coaster speeds
Danger_Drop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
RailGun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
    
# Roller coaster length arangement para la velocidad
length = np.arange(0, 130, 10)
```

```python
%matplotlib notebook
```

```python
#Graph
dd = plt.plot(length, Danger_Drop, label = "Danger Drop", color = "r")
rg = plt.plot(length, RailGun, label = "RailGun", color = "g")

plt.legend(loc = "upper right")
plt.title(" Roller Coaster Speed Over Time")
plt.xlabel("Runtime [10s]")
plt.ylabel("Speed [MPH]")
plt.grid()
```

```python
# Graphing a bar chart, only good for univariate data
users = [1500, 3000, 5200, 6400, 4300]
x_ax = np.arange(len(users))
```

```python
%matplotlib notebook
```

```python

plt.bar(x_ax, users, color = "pink", label = "Programming Language", align = "center")
plt.legend(loc = "upper right")
```

```python
tick_locations = [value for value in x_ax]
plt.xticks(tick_locations, ["Java", "C++", "Python", "Ruby", "Clojure"])

plt.xlim(-0.75, 0.5, 0.0)
```
